```python
# For outliers

Q1 = df['ticket_price_cr'].quantile(0.25)
Q3 = df['ticket_price_cr'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)
```

```python
#replace outlier as NaN
df['ticket_price_cr'] = df['ticket_price_cr'].apply(
    lambda x: x if (x >= lower_limit and x <= upper_limit) else np.nan
)
```

```python
df['ticket_price_cr'].isnull().sum()
```

```python
df['ticket_price_cr'] = df.groupby(
    ['configuration', 'micro_market', 'developer_name']
)['ticket_price_cr'].transform(lambda x: x.fillna(x.mean()))
```
